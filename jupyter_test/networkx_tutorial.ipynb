```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge(1,2)
G.add_edge(2,3, weight=0.9)

nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
G = nx.complete_graph(5)
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
edge_list = [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)]
G = nx.Graph()
G = nx.from_edgelist(edge_list)
nx.draw_planar (G, with_labels=True)
plt.show()
```

```python
edge_list = [(1, 2), (2, 3), (3, 4), (3, 5), (4, 6), (6, 7)]
G = nx.from_edgelist(edge_list)

nx.draw_spring(G, with_labels=True)
plt.show()

nx.draw_circular(G, with_labels=True)
plt.show()

nx.draw_shell(G, with_labels=True)
plt.show()

nx.draw_spectral(G, with_labels=True)
plt.show()

nx.draw_random(G, with_labels=True)
plt.show()

nx.draw_planar (G, with_labels=True)
plt.show()
```

```python
print(nx.adjacency_matrix(
    G
))
```

```python
G = nx.from_numpy_array(np.array([[0, 1, 0],
                                  [1, 1, 1],
                                  [0, 0, 0]]))

""" Interpretacin: """
"""  """
""" Nodo 0: conectado con nodo 1 (valor 1). """
""" Nodo 1: conectado con nodo 0 (1), consigo mismo (1), y con nodo 2 (1). """
""" Nodo 2: no tiene salidas (todos ceros). """

nx.draw_spring(G, with_labels=True)
plt.show()
print(nx.adjacency_matrix(
    G
))
```
